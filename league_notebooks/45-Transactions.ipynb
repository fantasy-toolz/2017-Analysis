```python

invite = {}
f = open('data/translog0102218.dat')

for line in f:
    if ('invite' in line) & ('contract' not in line):
        name = line.split()[-5:-3]
        team = line.split()[0:2]
    if ('invite' in line) & ('contract' in line):
        name = line.split()[-13:-11]
        team = line.split()[0:2]
        #invite[]
    

    invite[name[0]+' '+name[1]] = team[0]+' '+team[1]


for key in invite.keys():
    print key,':',invite[key]


    
    
```

```python



dl = {}
f = open('data/translog033118.dat')

for line in f:
    if ('disabled list' in line):
        #print line
        
        try:
            w = line.split().index('placed')
            name = line.split()[w+2:w+4]
            #print name
            team = line.split()[0:w]

            dl[name[0]+' '+name[1]] = team[0]+' '+team[1]

        except:
            pass
        
        try:
            w = line.split().index('transferred')
            name = line.split()[w+2:w+4]
            #print name
            team = line.split()[0:w]

            dl[name[0]+' '+name[1]] = team[0]+' '+team[1]

        except:
            pass
            
for key in dl.keys():
    print key,':',dl[key]


    
    
```
