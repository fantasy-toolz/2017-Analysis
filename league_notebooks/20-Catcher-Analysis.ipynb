```python
# boilerplate imports

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import csv
%matplotlib inline
```

```python


def read_batting_data(infile):
    BatData = np.genfromtxt(("\t".join(i) for i in csv.reader(open(infile))), delimiter="\t",\
                  dtype={'names': ("Name","Team","G","AB","PA","H",\
                                   "1B","2B","3B","HR","R","RBI","BB","IBB",
                                   "SO","HBP","SF","SH","GDP","SB","CS","AVG","playerid"),\
                         'formats': ('S20','S20','i2','f4','f4','f4',\
                                     'f4','f4','f4','f4','f4','f4','f4','f4',\
                                    'f4','f4','f4','f4','f4','f4','f4','f4','i2')},\
                 skip_header=1)
    return BatData


# 70 PA+ only
Catchers   = read_batting_data('../batting_notebooks/2017Batting/Catchers_071617.csv')

```

```python
pname = Catchers['Name']
rbis  = Catchers['RBI']

fig = plt.figure(figsize=(3,1))

plt.subplots_adjust(hspace=0.001)

ax = fig.add_subplot(111)

howmany = 20

ax.plot(np.linspace(1.,howmany,howmany),rbis[(-1.*rbis).argsort()[0:howmany]])
ax.set_ylabel('RBIs',size=12)
ax.axis([1.,howmany,np.min(rbis),np.max(rbis)])
_ = ax.set_xticks(np.linspace(1.,howmany,howmany))


ax2 = ax.twiny()
_ = ax2.set_xticks(np.arange(0,howmany+5,5))
_ = ax2.set_xticklabels(np.arange(0,howmany+4,5)+1, rotation='vertical',size=8)


#ax.set_title('Top Catchers, 2017')

# You can specify a rotation for the tick labels in degrees or with keywords.
_ = ax.set_xticks(np.linspace(1.,howmany,howmany))
_ = ax.set_xticklabels(pname[(-1.*rbis).argsort()], rotation='vertical',size=8)
```

### Average

This looks like the easiest market inefficiency to exploit (and was my original plan). With the exception of Posey and Realmuto, none of the top guys are tough to get.

However, the shape of this curve is intriguing to me. There's a floor for catcher average, and with low PAs, they really shouldn't swing it one way or another. I'm going to say that these guys are almost always better to own that having no one, unless your league is insanely shallow and you can get a top 100 pitcher for that spot.

```python
pname = Catchers['Name']
rbis  = Catchers['SB']

fig = plt.figure(figsize=(3,1))

plt.subplots_adjust(hspace=0.001)

ax = fig.add_subplot(111)

howmany = 20

ax.plot(np.linspace(1.,howmany,howmany),rbis[(-1.*rbis).argsort()[0:howmany]])
ax.set_ylabel('SBs',size=12)
ax.axis([1.,howmany,.25,np.max(rbis)])
_ = ax.set_xticks(np.linspace(1.,howmany,howmany))


ax2 = ax.twiny()
_ = ax2.set_xticks(np.arange(0,howmany+5,5))
_ = ax2.set_xticklabels(np.arange(0,howmany+4,5)+1, rotation='vertical',size=8)


#ax.set_title('Top Catchers, 2017')

# You can specify a rotation for the tick labels in degrees or with keywords.
_ = ax.set_xticks(np.linspace(1.,howmany,howmany))
_ = ax.set_xticklabels(pname[(-1.*rbis).argsort()], rotation='vertical',size=8)
```
