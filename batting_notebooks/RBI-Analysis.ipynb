As a first plot, how did teams do in recording RBIs last year?

```python
team_rbis = Teams2016['RBI']
team_name = Teams2016['Team']

plt.figure(figsize=(5,2))
plt.plot(np.linspace(1.,team_rbis.size,team_rbis.size),team_rbis[(-1.*team_rbis).argsort()])
plt.ylabel('RBIs',size=24)
plt.title('Team Total RBIs, 2016')
_ = plt.axis([1.,30.,np.min(team_rbis),np.max(team_rbis)])

# You can specify a rotation for the tick labels in degrees or with keywords.
_ = plt.xticks(np.linspace(1,30,30), team_name[(-1.*team_rbis).argsort()], rotation='vertical')
```

The team rankings makes me think that we'll want to put some sort of caveat on looking at individual stats based on the team they are on. Let's start looking at individual players by setting up a ranking within the league.

```python
rbis = Total2016['RBI']
pas = Total2016['PA']
names = Total2016['Name']

plt.figure(figsize=(5,2))
plt.plot(np.linspace(1.,rbis.size,rbis.size),rbis[(-1.*rbis).argsort()])
plt.xlabel('Player Rank',size=24)
plt.ylabel('RBIs',size=24)
plt.title('Top 320 RBI production curve, 2016')
_ = plt.axis([1.,320.,np.min(rbis),np.max(rbis)])
```

#### What to notice? 

First, you REALLY want those guys who are bending the curve upward above 100 RBIs. The fact that the slope changes indicates that they are exponentially more valuable than guys in the shallower slope set. They are a pretty typical list:



```python
print('100+ RBIs, 2016')
print('-----------------')
for indx in (-1.*rbis).argsort()[0:22]:
    print('{0:4d} -- {1:20s} ({2:3d})'.format(int(rbis[indx]),names[indx],int(pas[indx])))
```

The last column here is the number of plate appearances. Obviously, all these guys had a full season's worth of plate appearances, with the only guy under 600 PAs being Daniel Murphy.

This top RBI-producer trend persists if you look at this as RBI/PA--no Gary Sanchez-type outliers here (see the HR analysis for context).

The next question to work on is **how** do guys get RBIs?



There must be some effect with batting order. Therefore, let's start by looking at the RBI total as a function of batting order (restricting to players with >100 PAs at that batting order position).

```python
total_list = np.array([First2016,Second2016,Third2016,Fourth2016,Fifth2016,Sixth2016,Seventh2016,Eighth2016,Ninth2016])

total_order_rbis = np.array([np.sum(x['RBI']) for x in total_list])
total_order_pas = np.array([np.sum(x['PA']) for x in total_list])



fig = plt.figure(figsize=(3,2))

plt.subplots_adjust(hspace=0.001)

ax = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax.plot(np.linspace(1.,9.,9),total_order_rbis)
ax.set_ylabel('Total RBIs')
ax.set_xticklabels(())

ax2.plot(np.linspace(1.,9.,9),total_order_rbis/total_order_pas)
ax2.set_ylabel('RBI/PA')
ax2.set_xlabel('Batting Order Position')

```

So, while the most *efficient* batting order position is 4th, 3rd and 4th end up with almost the same total number of RBIs.

Also note that the RBI/PA variance is not that large (9-16%), but the total number of RBIs has a huge range. In part, this is due to the cycling of the 8th and 9th positions in the batting order (and the limit to >100 PAs). This, to me, is saying that if a guy hits *consistently* in the 8th spot, you'll take a hit with the PAs, but the RBI/PA won't kill you--they still have some utility.

To take this one step further, I'm curious about whether there is any variance in batting order for a specific player--e.g. does Nolan Arenado hit more efficiently for RBIs in one lineup spot or another?

I'll do this by looking at the RBI producers with over 100 RBIs in 2016, and then break them down by where they appeared in the batting order for over 100 PAs in 2016.

```python
total_list = np.array([First2016,Second2016,Third2016,Fourth2016,Fifth2016,Sixth2016,Seventh2016,Eighth2016,Ninth2016])

fig = plt.figure(figsize=(3,3))

# search the dictionaries for players
pnames = names[(-1.*rbis).argsort()[0:22]]

for playername in pnames:
    
    order = []
    rbipa = []

    for indx,val in enumerate(total_list):
    
        if playername in val['Name']:
            playerval = np.where(playername==val['Name'])[0]
            
            order.append(indx+1)
            rbipa.append(float(val['RBI'][playerval]/val['PA'][playerval]))

            # uncomment this for the raw names and numbers.
            #print('{0:>20s} batted {1:1d}, with {2:3d} RBIs in {3:3d} plate appearances ({4:4.3f}%).'\
            #.format(playername,indx+1,int(val['RBI'][playerval]),\
            #        int(val['PA'][playerval]),float(val['RBI'][playerval]/val['PA'][playerval])))
            
    try:      
        plt.plot(order,rbipa,color='black')
        if rbipa[-1] > 0.19:
            plt.text(order[-1],rbipa[-1],playername,size=6)
    except:
        pass
    
plt.axis([0.,7.,0.1,0.25])
_ = plt.xlabel('Batting Order Position',size=18)
_ = plt.ylabel('RBI/PA',size=18)
```

The upward slope of the majority of lines suggests that for individual players, it is better to be further down in the lineup. For instance, Adam Duvall is a much more efficient RBI machine when he moves from 4th to 5th in the lineup--despite an overall falloff in league efficiency. The same is true for Mark Trumbo.

And that guy who got RBIs in the leadoff spot? That's Mookie Betts, but you'd definitely prefer him to be batting 4th for your RBI totals.
